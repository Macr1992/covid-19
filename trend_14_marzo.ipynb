```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data  = pd.read_csv("data.csv")
```

```python
t = data["time"]
logn = np.log(data["count"])

m,b = np.polyfit(t, logn, 1)

x = np.linspace(t.min(), t.max())
y = m*x + b
plt.semilogy(data["time"], data["count"], 'o')
plt.plot(x, np.exp(y))
plt.title('Doubling Time: {:.2f} days'.format(1.0/m))
plt.xlabel('Time [Days]')
plt.ylabel('Cumulative number of infections')
```

```python
x = np.linspace(t.min(), 20)
y = m*x + b
plt.semilogy(data["time"], data["count"], 'o')
plt.plot(x, np.exp(y))
plt.title('Doubling Time: {:.2f} days'.format(1.0/m))
plt.xlabel('Time [Days]')
plt.ylabel('Cumulative number of infections')
```

```python
# predictions
for i in range(20):
    print("{} de Marzo \t {} casos".format(9+i, int(np.exp(m*i+b))))
```
